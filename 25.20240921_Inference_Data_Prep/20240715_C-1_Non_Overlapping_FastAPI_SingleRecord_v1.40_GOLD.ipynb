```python
import requests
import numpy as np
import pandas as pd

# Parameters
data_dir = '/home/ubuntu/efs-w210-capstone-ebs/04A.Local_Data_Files'
file_prefix = "06.20240714_062624_non_overlap_full_test"  # Replace with your actual file prefix

# Load the test data
X_test = np.load(f"{data_dir}/{file_prefix}_X_test.npy")
y_test = np.load(f"{data_dir}/{file_prefix}_y_test.npy")

# Drop the first column from X_test
X_test = X_test[:, :, 1:]

# Take the first record from X_test
sample_input = X_test[0].tolist()  # Convert to list for JSON serialization

# Make a POST request to the API
response = requests.post('http://localhost:8000/predict', json={'sequence': sample_input})

# Print the result
api_response = response.json()
# print("API Response:")
# print(api_response)

# Print the actual label for comparison
# print("\nActual label:")
# print(y_test[0])

# Print the shape of the input for verification
# print("\nShape of input sequence:")
# print(np.array(sample_input).shape)

# Load the original parquet file with sequence mapping
input_file = f"{data_dir}/03.20240715_143154_orig_input_w_seq_mapping_master_FINAL.parquet"
original_df = pd.read_parquet(input_file)

# Ensure Seq_Num is integer for proper merging
original_df['Seq_Num'] = original_df['Seq_Num'].astype(int)

# Identify the first instance where Train_Test is 'Test'
seq_start = original_df[original_df['Train_Test'] == 'Test'].iloc[0]['Seq_Num']

# Calculate detailed sequence numbers
total_seq_start = seq_start
total_seq_end = total_seq_start + 23 - 1
source_seq_start = total_seq_start
source_seq_end = source_seq_start + 20 - 1
gap_seq_start = source_seq_end + 1
gap_seq_end = gap_seq_start + 1
prediction_seq_start = total_seq_end - 1
prediction_seq_end = total_seq_end

# Extract the relevant row from the original dataframe
additional_info = original_df[original_df['Seq_Num'] == source_seq_start].iloc[0]

# Create a DataFrame for displaying the results
results_df = pd.DataFrame({
    "Total_Seq_Start": [total_seq_start],
    "Total_Seq_End": [total_seq_end],
    "Source_Seq_Start": [source_seq_start],
    "Source_Seq_End": [source_seq_end],
    "Gap_Seq_Start": [gap_seq_start],
    "Gap_Seq_End": [gap_seq_end],
    "Prediction_Seq_Start": [prediction_seq_start],
    "Prediction_Seq_End": [prediction_seq_end],
    "raw_output": [api_response['raw_output']],
    "sigmoid_output": [api_response['sigmoid_output']],
    "Predicted": [api_response['thresholded_prediction']],
    "Actual": [y_test[0]]
})

# Convert Source_Seq_Start to integer for proper merging
results_df['Source_Seq_Start'] = results_df['Source_Seq_Start'].astype(int)

# Merge with additional information
merged_df = results_df.merge(original_df, left_on='Source_Seq_Start', right_on='Seq_Num', how='left')

# Display the first 20 columns of the new DataFrame
pd.set_option('display.max_columns', 30)
# print("\nEnhanced Results:")
display(merged_df.iloc[:, :30])
```
