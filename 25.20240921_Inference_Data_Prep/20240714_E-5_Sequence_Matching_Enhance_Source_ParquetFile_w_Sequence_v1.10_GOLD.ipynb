# <font color = blue> Enhance Original Input File with Sequence Info

```python
import pandas as pd
from datetime import datetime

# Paths
data_dir = '/home/ubuntu/efs-w210-capstone-ebs/04A.Local_Data_Files'
input_file = f"{data_dir}/20240709_full_more_new_features_v1.00.parquet"
output_dir = '/home/ubuntu/efs-w210-capstone-ebs/04C.Local_Inference_Eval_Files'

# Load data
encoded_feature_input_file_df = pd.read_parquet(input_file)

# Create new columns
seq_num = encoded_feature_input_file_df.index
train_test = ['Train' if x <= 11614 else 'Test' for x in seq_num]

# Insert the new columns at the beginning
encoded_feature_input_file_df.insert(0, 'Seq_Num', seq_num)
encoded_feature_input_file_df.insert(1, 'Train_Test', train_test)

# Save the new DataFrame to a parquet file
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = f"{output_dir}/{timestamp}_orig_input_w_seq_mapping_master.parquet"

encoded_feature_input_file_df.to_parquet(output_file, index=False)

print(f"New file saved to {output_file}")

# Display the first 20 columns of the first few rows to verify
display(encoded_feature_input_file_df.iloc[:, :20].head())
```
