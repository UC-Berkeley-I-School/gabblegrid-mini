```python
import matplotlib.pyplot as plt

def draw_simple_lstm_diagram():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define the structure
    ax.text(0.1, 0.8, 'Input Layer\n(max_events + 4)', bbox=dict(facecolor='lightblue', edgecolor='black'), fontsize=12, ha='center')
    ax.text(0.4, 0.8, 'LSTM Layer\n(hidden_size, num_layers)', bbox=dict(facecolor='lightgreen', edgecolor='black'), fontsize=12, ha='center')
    ax.text(0.7, 0.8, 'Dropout Layer\n(0.5)', bbox=dict(facecolor='lightyellow', edgecolor='black'), fontsize=12, ha='center')
    ax.text(1.0, 0.8, 'Fully Connected Layer\n(output_size=1)', bbox=dict(facecolor='lightcoral', edgecolor='black'), fontsize=12, ha='center')
    ax.text(1.3, 0.8, 'Output Layer\n(Sigmoid Activation)', bbox=dict(facecolor='lightgray', edgecolor='black'), fontsize=12, ha='center')

    # Add arrows
    ax.annotate('', xy=(0.2, 0.8), xytext=(0.3, 0.8), arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate('', xy=(0.5, 0.8), xytext=(0.6, 0.8), arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate('', xy=(0.8, 0.8), xytext=(0.9, 0.8), arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate('', xy=(1.1, 0.8), xytext=(1.2, 0.8), arrowprops=dict(arrowstyle="->", lw=2))

    ax.axis('off')
    plt.title('LSTM Model Block Diagram')
    plt.show()

draw_simple_lstm_diagram()
```
